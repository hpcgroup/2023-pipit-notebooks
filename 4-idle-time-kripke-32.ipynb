```python
import pipit

from pipit.plotting import _util
_util.theme = "paper"
import pandas as pd
pd.set_option('max_colwidth', None)

dir = './kripke/ScoreP-Kripke-32/'
trace_32 = pipit.Trace.from_otf2(dir)
```

```python
idle_32 = trace_32.idle_time(idle_functions=['MPI_Testany', 'MPI_Allreduce'], MPI_events=True)
idle_32.sort_values(by=['Idle Time', 'Process'], ascending=False)
```

```python
trace_32.plot_timeline()
```

```python
df = trace_32.events
dff = df.loc[df['Timestamp (ns)'] > 1530000000].loc[df['Timestamp (ns)'] < 1910000000]
most_idle = dff.loc[dff['Process'] == 25]
least_idle = dff.loc[dff['Process'] == 16]
```

```python
def print_most_funcs(df):
    df = df.groupby(['Name'])['time.inc'].sum().reset_index(name='Time').sort_values(['Time'], ascending=False)
    display(df.head())
```

```python
print_most_funcs(most_idle)
print_most_funcs(least_idle)
```
